## Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?


> A **probability distribution** is a function that describes the possible values of a random variable and their corresponding probabilities. It is a mathematical description of how likely a particular outcome is. <br>
> For example, the **probability distribution** for rolling a die is:

            Outcome | Probability
            ------- | --------
            1 | 1/6
            2 | 1/6
            3 | 1/6
            4 | 1/6
            5 | 1/6
            6 | 1/6

> This **distribution** tells us that each of the six possible outcomes (1, 2, 3, 4, 5, or 6) is equally likely. <br>
> **Probability distributions** can be used to predict the likelihood of future events. For example, if we know the probability distribution for rolling a die, we can predict that the probability of rolling a 6 is 1/6.

## Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

> Yes, there is a distinction between true random numbers and pseudo-random numbers.

> **True random numbers** are generated by physical processes that are inherently unpredictable. For example, the decay of radioactive atoms is a true random process. <br>
> **Pseudo-random numbers** are generated by algorithms that produce sequences of numbers that appear to be random. However, these sequences are actually deterministic, meaning that they can be predicted if you know the algorithm and the initial state.

> <h4><b><i>Why are pseudo-random numbers considered “good enough” for most applications?</i></b></h4>

>> **Pseudo-random numbers** are considered "good enough" for most applications because they are very difficult to distinguish from true random numbers. <br>
>> Here are some of the reasons why **pseudo-random numbers** are considered "good enough" for most applications:
>> * They are very efficient to generate.
>> * They are very difficult to distinguish from true random numbers.
>> * They are sufficient for most applications.

> However, it is important to note that **pseudo-random numbers** are not perfect. They can be predictable if you know the algorithm and the initial state. Therefore, it is important to choose a pseudo-random number generator that is well-designed and that has been tested for security.

## Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

> The two main factors that influence the behavior of a normal probability distribution are **mean** and **standard deviation**.
> * **Mean** is the average value of the distribution. It is the point around which the distribution is centered.
> * **Standard deviatio**n is a measure of how spread out the distribution is. It is the square root of the variance, which is a measure of how far the values in the distribution are from the mean.

> The **mean** and **standard deviation** of a normal probability distribution determine the shape of the distribution. The **mean** determines the center of the distribution, and the **standard deviation** determines how spread out the distribution is.

## Q4. Provide a real-life example of a normal distribution.


> Some real-life examples of a normal distribution are:
> * **The heights of people**: The heights of people are normally distributed, with the mean being the average height of the population and the standard deviation being the typical variation in height. This means that most people will be within 1 standard deviation of the mean, and a few people will be much taller or shorter than the mean.
> * **The scores on tests**: The scores on tests are also normally distributed, with the mean being the average score and the standard deviation being the typical variation in scores. This means that most people will score within 1 standard deviation of the mean, and a few people will score much higher or lower than the mean.
> * **The number of heads in 10 coin flips**: If you flip a coin 10 times, the number of heads you get will be normally distributed, with the mean being 5 heads and the standard deviation being 2 heads. This means that most of the time, you will get between 3 and 7 heads, and only rarely will you get more than 7 heads or less than 3 heads.

## Q8. What is the relationship between exponentiation and logarithms?

> **Exponentiation** and **logarithms** are inverse functions of each other. This means that the logarithm of a number is the exponent to which another number must be raised to equal the first number.

> For example, the logarithm of 100 to the base 10 is 2, because 10 raised to the power of 2 is 100.: ```log10(100) = 2``` <br>
> The inverse of this equation is: ```10^2 = 100```



## Q9. What are the three logarithmic functions that Python supports?

> Python supports three logarithmic functions are:
> * ***log(x)***: This is the natural logarithm of x, also known as ln(x).
> * ***log10(x)***: This is the logarithm of x to the base 10.
> * ***log2(x)***: This is the logarithm of x to the base 2.


````python
import math

x = 1000

natural_log = math.log(x)  # Natural logarithm of x (base e)
log_base_10 = math.log10(x)  # Logarithm of x to base 10
log_base_2 = math.log2(x)  # Logarithm of x to base 2

print(natural_log)  # Output: 6.907755278982137
print(log_base_10)  # Output: 3.0
print(log_base_2)  # Output: 9.965784284662087
````
