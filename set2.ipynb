## Question: A website requires the users to input username and password to register. Write a program to **check the validity of password input by users.**
        Following are the criteria for checking the password:
            1. At least 1 letter between [a-z]
            2. At least 1 number between [0-9]
            1. At least 1 letter between [A-Z]
            3. At least 1 character from [$#@]
            4. Minimum length of transaction password: 6
            5. Maximum length of transaction password: 12
            
                    Your program should accept a sequence of comma separated passwords and will check them
                    according to the above criteria. Passwords that match the criteria are to be printed, each
                    separated by a comma.
                    Example
                    If the following passwords are given as input to the program:
                    ABd1234@1,a F1#,2w3E*,2We3345
                    Then, the output of the program should be:
                    ABd1234@1
```python
import re


def is_valid_password(password):
    criteria = [
          re.compile("[a-z]").search(password),
          re.compile("[0-9]").search(password),
          re.compile("[A-Z]").search(password),
          re.compile("[$#@]").search(password),  ]
    
    if len(password) < 6 or len(password) > 12:
        return False
    for c in criteria:
        if not c:
            return False
    return True


passwords =  ["ABd1234@1","aghhhhhhkkhhhhgggggg", "F1#","2nw3E*","2We3345"]

valid_passwords = []
for i in passwords:
    if is_valid_password(i):
        valid_passwords.append(i)
print("Valid passwords: ", ",".join(valid_passwords))
```

# Question 6: Please write a **binary search** function which searches an item in a sorted list. The function should return the index of element to be searched in the list.
```python
def binary_search(list, item):
    low = 0
    high = len(list) - 1
    while low <= high:
        mid = (low + high) // 2
        if list[mid] == item:
            return mid
        elif list[mid] < item:
            low = mid + 1
        else:
            high = mid - 1
    return -1


my_list = [1, 3, 5, 7, 9]
search_item = 5

index = binary_search(my_list, search_item)

if index == -1:
    print("The item is not found.")
else:
    print("The item is found at index:", index)
```

## Question: **The Fibonacci Sequence** is computed based on the following formula:
            f(n)=0 if n=0
            f(n)=1 if n=1
            f(n)=f(n-1)+f(n-2) if n>1
            
### Please write a program using list comprehension to print the Fibonacci Sequence in comma separated form with a given n input by console.
            Example:
            If the following n is given as input to the program:
            7
            Then, the output of the program should be:
            0,1,1,2,3,5,8,13
```python
def fibonacci(n):
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        fibs = [0, 1]
    for i in range(2, n + 1):
        fibs.append(fibs[i - 1] + fibs[i - 2])
    return fibs

num = int(input("Enter the terms for fibonacci series: "))

print(",".join([str(f) for f in fibonacci(num)]))
```

## Question: Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10) 010101001(2) = 1 + 8 + 32 + 128. 
### Going from right to left, the value of the most right bit is 1, now from that every bit to the left will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1).

                Examples
                binary(1) â "1"
                # 1*1 = 1
                binary(5) â "101"
                # 1*1 + 1*4 = 5
                binary(10) â "1010"
                # 1*2 + 1*8 = 10
```python
def binary(decimal):
    if decimal == 0:
        return "0"

    binary_string = ""
    while decimal > 0:
        remainder = decimal % 2
        binary_string = str(remainder) + binary_string
        decimal = decimal // 2

    return binary_string

print(binary(1))  
print(binary(5))   
print(binary(710))  
```

## Question: Create a function that returns True if a given inequality expression is correct and False otherwise.

            correct_signs("3 < 7 < 11") â True
            correct_signs("13 > 44 > 33 > 1") â False
            correct_signs("1 < 2 < 6 < 9 > 3") â True
```python
def correct_signs(expression):
    tokens = expression.split()
    for i in range(1, len(tokens), 2):
        operator = tokens[i]
        num1 = int(tokens[i - 1])
        num2 = int(tokens[i + 1])
        
        if (operator == "<" and num1 >= num2) or (operator == ">" and num1 <= num2):  # Check if the inequality is not satisfied
            return False
        
    return True

print('correct_signs("3 < 7 < 11") ==>', correct_signs("3 < 7 < 11"))
print('correct_signs("13 > 44 > 33 > 1") ==>', correct_signs("13 > 44 > 33 > 1"))
print('correct_signs("1 < 2 < 6 < 9 > 3") ==>' , correct_signs("1 < 2 < 6 < 9 > 3"))
```

## Question: Hamming distance is the number of characters that differ between two strings.
            To illustrate:
            String1: "abcbba"
            String2: "abcbda"
            Hamming Distance: 1 - "b" vs. "d" is the only difference.
### Create a function that computes the hamming distance between two strings.
            Examples
            hamming_distance("abcde", "bcdef") â 5
            hamming_distance("abcde", "abcde") â 0
            hamming_distance("strong", "strung") â 1
```python
def hamming_distance(string1, string2):
    if len(string1) != len(string2):
        raise ValueError("Strings must be of equal length")
    distance = 0
    for i in range(len(string1)):
        if string1[i] != string2[i]:
            distance += 1
    return distance

print('hamming_distance("abcde", "bcdef") ==>', hamming_distance("abcde", "bcdef"))
print('hamming_distance("abcde", "abcde") ==>', hamming_distance("abcde", "abcde"))
print('hamming_distance("strong", "strung") ==>', hamming_distance("strong", "strung"))
```

            Examples
            reverse("Hello World") â "DLROw OLLEh"
            reverse("ReVeRsE") â "eSrEvEr"
            reverse("Radar") â "RADAr"
```python
def reverse(string):
    new_string = ""
    for char in string:
        if char.isupper():
            new_char = char.lower()
        else:
            new_char = char.upper()
        new_string += new_char
    return new_string[::-1]

print(f'reverse("Hello World") ==> "{reverse("Hello World")}"')
print(f'reverse("ReVeRsE") ==> "{reverse("ReVeRsE")}"')
print(f'reverse("Radar") ==> "{reverse("Radar")}"')
```

            move_to_end(["a", "a", "a", "b"], "a") â ["b", "a", "a", "a"]
```python
def move_to_end(lst, element):    
    count = lst.count(element)              
    lst = [x for x in lst if x != element]      
    lst.extend([element] * count)           
    
    return lst

print('move_to_end([1, 3, 2, 4, 4, 1], 1) ==>', move_to_end([1, 3, 2, 4, 4, 1], 1))
print('move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ==>', move_to_end([7, 8, 9, 1, 2, 3, 4], 9))
print('move_to_end(["a", "a", "a", "b"], "a") ==>', move_to_end(["a", "a", "a", "b"], "a"))
```

```python
def dict_to_list(dictionary):
    list_of_tuples = []
    for key, value in dictionary.items():
        tuples = (key, value)
        list_of_tuples.append(tuples)
    list_of_tuples.sort()
    return list_of_tuples


print(dict_to_list({
                "D": 1,
                "B": 2,
                "C": 3
                }))

print(dict_to_list({
                "likes": 2,
                "dislikes": 3,
                "followers": 10
                }))
```

## Question: Write a function that creates a dictionary with each (key, value) pair being the (lower case, upper case) versions of a letter, respectively.
        Examples
        mapping(["p", "s"]) â { "p": "P", "s": "S" }
        mapping(["a", "b", "c"]) â { "a": "A", "b": "B", "c": "C" }
        mapping(["a", "v", "y", "z"]) â { "a": "A", "v": "V", "y": "Y", "z": "Z" }
> <b><u> Notes </u></b>
>> All of the letters in the input list will always be lowercase.
```python
def mapping(lst):
    dictionary = {}
    for char in lst:
        lowercase = char
        uppercase = char.upper()
        dictionary[lowercase] = uppercase
    return dictionary

print('mapping(["p", "s"]) ==>', mapping(["p", "s"]))
print('mapping(["a", "b", "c"]) ==>', mapping(["a", "b", "c"]))
print('mapping(["a", "v", "y", "z"]) ==>', mapping(["a", "v", "y", "z"]))
```

## Question: Write a function, that replaces all vowels in a string with a specified vowel.
            Examples
            vow_replace("apples and bananas", "u") â "upplus und bununus"
            vow_replace("cheese casserole", "o") â "chooso cossorolo"
            vow_replace("stuffed jalapeno poppers", "e") â "steffed jelepene peppers"
> <b><u> Notes </u></b>
>> All words will be lowercase. Y is not considered a vowel.
```python
def vow_replace(words, vowel):
    vowels = "aeiou"
    new_string = ""
    for i in words:
        if i in vowels:
            new_string += vowel
        else:
            new_string += i
    return new_string

print(f'vow_replace("apples and bananas", "u") ==> "{vow_replace("apples and bananas", "u")}"')
print(f'vow_replace("cheese casserole", "o") ==> "{vow_replace("cheese casserole", "o")}"')
print(f'vow_replace("stuffed jalapeno poppers", "e") ==> "{vow_replace("stuffed jalapeno poppers", "e")}"')
```

## Question: Create a function that takes a string as input and capitalizes a letter if its ASCII code is even and returns its lower case version if its ASCII code is odd.
            Examples
            ascii_capitalize("to be or not to be!") â "To Be oR NoT To Be!"
            ascii_capitalize("THE LITTLE MERMAID") â "THe LiTTLe meRmaiD"
            ascii_capitalize("Oh what a beautiful morning.") â "oH wHaT a BeauTiFuL moRNiNg."
```python
def ascii_capitalize(string):
    new_string = ""
    for letter in string:
        ascii_code = ord(letter)
        if ascii_code % 2 == 0:
            new_string += letter.upper()
        else:
            new_string += letter.lower()
    return new_string

print(f'ascii_capitalize("to be or not to be!") ==> "{ascii_capitalize("to be or not to be!")}"')
print(f'ascii_capitalize("THE LITTLE MERMAID") ==> "{ascii_capitalize("THE LITTLE MERMAID")}"')
print(f'ascii_capitalize("Oh what a beautiful morning.") ==> "{ascii_capitalize("Oh what a beautiful morning.")}"')
```            
---
---
